```python
# === Bloque 1: Preparación Total de Datos ===
import mne
from mne.preprocessing import ICA
import os
import urllib.request

print("--- Preparando el entorno de análisis completo ---")

# 1. Asegurar que el archivo de datos existe
url = 'https://physionet.org/files/chbmit/1.0.0/chb01/chb01_01.edf'
filename = 'chb01_01.edf'
if not os.path.exists(filename):
    print(f"Descargando '{filename}'...")
    urllib.request.urlretrieve(url, filename)
else:
    print(f"El archivo '{filename}' ya existe.")

# 2. Cargar y filtrar los datos
print("Cargando y filtrando datos...")
raw = mne.io.read_raw_edf(filename, preload=True, verbose=False)
raw_filtered = raw.copy().filter(l_freq=1.0, h_freq=40.0, verbose=False)

# 3. Entrenar ICA y definir componentes a excluir
print("Entrenando ICA para limpiar artefactos...")
ica = ICA(n_components=15, max_iter='auto', random_state=97)
ica.fit(raw_filtered, verbose=False)
ica.exclude = [0, 1] # Excluimos los componentes de parpadeo que identificamos

# 4. Aplicar la limpieza y crear la variable final
raw_cleaned = ica.apply(raw_filtered.copy(), verbose=False)

print("\n¡Éxito! La variable 'raw_cleaned' con los datos limpios está lista.")
```

```python
# === Bloque 3: Cálculo de Entropía de Permutación ===
import numpy as np
import antropy as ant

datos_eeg = raw_cleaned.get_data(units='V')
nombres_canales = raw_cleaned.ch_names
print("\nCalculando la Entropía de Permutación para cada canal...")

for i in range(len(nombres_canales)):
    entropia = ant.perm_entropy(datos_eeg[i], order=3, delay=1, normalize=True)
    print(f"Canal: {nombres_canales[i]:<10} | Entropía de Permutación: {entropia:.4f}")

print("\nCálculo de entropía finalizado.")
```

```python
# === Bloque 4 (Versión a Prueba de Fallos): Visualización Topográfica ===

import matplotlib.pyplot as plt
import numpy as np
import antropy as ant
import mne

# --- Pasos 1 y 2: Calcular y promediar la entropía (sin cambios) ---
datos_eeg = raw_cleaned.get_data(units='V')
nombres_canales = raw_cleaned.ch_names
entropias = {name: ant.perm_entropy(datos_eeg[i], normalize=True) for i, name in enumerate(nombres_canales)}
entropia_promediada = {}
conteo_canales = {}
for nombre_bipolar, valor_entropia in entropias.items():
    nombre_principal = nombre_bipolar.split('-')[0]
    if nombre_principal not in entropia_promediada:
        entropia_promediada[nombre_principal] = 0
        conteo_canales[nombre_principal] = 0
    entropia_promediada[nombre_principal] += valor_entropia
    conteo_canales[nombre_principal] += 1
for nombre, suma in entropia_promediada.items():
    entropia_promediada[nombre] = suma / conteo_canales[nombre]

# --- Paso 3: Preparar datos y filtrar canales desconocidos (NUEVA LÓGICA) ---
nombres_originales = list(entropia_promediada.keys())
valores_originales = list(entropia_promediada.values())

# Corregimos mayúsculas
nombres_corregidos = [name.replace('FP', 'Fp') for name in nombres_originales]

# Obtenemos el mapa estándar
montage = mne.channels.make_standard_montage('standard_1020')

# FILTRO DE SEGURIDAD: Nos quedamos solo con los canales que existen en el montaje
canales_para_plot = []
valores_para_plot = []
for i, name in enumerate(nombres_corregidos):
    if name in montage.ch_names:
        canales_para_plot.append(name)
        valores_para_plot.append(valores_originales[i])

print(f"Se graficarán {len(canales_para_plot)} de {len(nombres_corregidos)} canales con posiciones conocidas.")

# Creamos la estructura de información solo con los canales válidos
info = mne.create_info(ch_names=canales_para_plot, sfreq=raw_cleaned.info['sfreq'], ch_types='eeg')
info.set_montage(montage)

# --- Paso 4: Dibujar el mapa ---
print("\nGenerando mapa topográfico de la entropía...")
fig, ax = plt.subplots(figsize=(6, 6))
im, cn = mne.viz.plot_topomap(data=valores_para_plot, pos=info, axes=ax, show=True,
                                cmap='viridis',
                                names=canales_para_plot)

# Añadimos una barra de color
cbar = plt.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
cbar.set_label('Entropía de Permutación (Normalizada)')
ax.set_title('Mapa Topográfico de Entropía', fontsize=16)

plt.show()
```

```python
# === Bloque 5: Gráfico de Barras de la Entropía por Canal ===

import matplotlib.pyplot as plt
import numpy as np

# 'entropia_promediada' ya existe en la memoria de nuestro cálculo anterior.
# Vamos a ordenar los canales de menor a mayor entropía para una mejor visualización.
sorted_items = sorted(entropia_promediada.items(), key=lambda item: item[1])
sorted_channels = [item[0] for item in sorted_items]
sorted_entropies = [item[1] for item in sorted_items]

# Creamos la figura para el gráfico
fig, ax = plt.subplots(figsize=(12, 7))

# Creamos el gráfico de barras
ax.bar(sorted_channels, sorted_entropies, color='skyblue')

# Añadimos títulos y etiquetas para mayor claridad
ax.set_ylabel('Entropía de Permutación (Normalizada)', fontsize=12)
ax.set_title('Entropía de Permutación por Canal (Ordenada)', fontsize=16)
ax.tick_params(axis='x', rotation=90) # Rotamos los nombres de los canales para que no se superpongan
ax.set_ylim(bottom=0.7) # Ajustamos el eje Y para resaltar las diferencias

# Mostramos el gráfico
plt.tight_layout() # Ajusta el layout para que todo sea visible
plt.show()
```

```python
# === Bloque 6: Construcción de la Red - Matriz de Correlación ===
import numpy as np
import matplotlib.pyplot as plt

print("Calculando la matriz de correlación entre todos los canales...")

# Extraemos los datos limpios. La transpuesta (.T) es necesaria
# para que las columnas sean los canales y las filas el tiempo.
datos_eeg = raw_cleaned.get_data().T

# Calculamos la correlación de Pearson.
# np.corrcoef crea una matriz donde cada elemento (i, j)
# es la correlación entre el canal i y el canal j.
matriz_correlacion = np.corrcoef(datos_eeg, rowvar=False)

print("Matriz de correlación calculada.")
print("Dimensiones de la matriz:", matriz_correlacion.shape)

# --- Visualización de la Matriz ---
print("\nMostrando la matriz de correlación como un mapa de calor...")
fig, ax = plt.subplots(figsize=(8, 8))
im = ax.imshow(matriz_correlacion, cmap='jet', interpolation='nearest')

# Añadimos la barra de color
cbar = fig.colorbar(im, ax=ax)
cbar.set_label('Correlación de Pearson')
ax.set_title('Matriz de Conectividad Funcional', fontsize=16)

plt.show()
```

```python
# === Bloque 7: Cálculo de Homología Persistente ===

import ripser
import numpy as np
import matplotlib.pyplot as plt

# La homología persistente funciona mejor con una matriz de 'distancia', no de 'similitud'.
# La convertimos: una correlación alta (1.0) es una distancia corta (0.0).
# Usamos el valor absoluto para que las anti-correlaciones fuertes también sean distancias cortas.
matriz_distancia = 1 - np.abs(matriz_correlacion)

print("Calculando la homología persistente de la red...")
# ripser.ripser calcula los 'diagramas de persistencia'
# 'maxdim=1' significa que buscaremos componentes (dimensión 0) y bucles (dimensión 1).
diagramas = ripser.ripser(matriz_distancia, maxdim=1, distance_matrix=True)['dgms']

persistencia_h0 = diagramas[0] # Betti 0: Componentes conectados
persistencia_h1 = diagramas[1] # Betti 1: Bucles o "agujeros" 1D

print(f"\nAnálisis completado.")
print(f"Se encontraron {len(persistencia_h1)} bucles unidimensionales (Betti 1) a lo largo de la filtración.")

# --- Visualización de la Persistencia ---
print("\nMostrando el diagrama de persistencia...")
# Este gráfico nos muestra cuándo 'nacen' y 'mueren' las características topológicas.
ripser.plot_dgms(diagramas, show=True)
plt.title('Diagrama de Persistencia de la Red Cerebral', fontsize=16)
plt.show()
```

```python
# === Bloque 9 (Versión 2 - Corregida y Simplificada): ===

import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import wasserstein_distance # <-- ¡LA SOLUCIÓN! USAMOS LA HERRAMIENTA OFICIAL

print("--- Iniciando Plan D (Versión Corregida): Cálculo directo de la curvatura ---")

# --- Parte 1: Nuestra función para calcular la Curvatura, ahora mucho más simple ---
def calcular_curvatura(G):
    print("Calculando curvatura para cada conexión...")
    for u, v in G.edges():
        # Grado de los nodos (número de vecinos)
        d_u = G.degree(u)
        d_v = G.degree(v)
        
        if d_u == 0 or d_v == 0:
            continue
            
        # Distribuciones de probabilidad uniformes sobre los vecinos
        p_u = np.ones(d_u) / d_u
        p_v = np.ones(d_v) / d_v
        
        # Usamos la función oficial de scipy para la distancia de transporte óptimo
        dist_transporte = wasserstein_distance(np.arange(d_u), np.arange(d_v), p_u, p_v)
        
        # Fórmula de la Curvatura de Ollivier-Ricci
        curvatura = 1 - dist_transporte
        
        # Guardamos el resultado en el grafo
        G[u][v]['ricciCurvature'] = curvatura
        
    return G

# --- Parte 2: Aplicamos la función a nuestros datos ---
matriz_correlacion_limpia = np.nan_to_num(matriz_correlacion)
G = nx.from_numpy_array(np.abs(matriz_correlacion_limpia))

G_con_curvatura = calcular_curvatura(G.copy())
print("Cálculo de curvatura finalizado.")


# --- Parte 3: Visualización (sin cambios) ---
print("\nGenerando el gráfico de la red coloreado por curvatura...")
curvaturas = [d['ricciCurvature'] for u, v, d in G_con_curvatura.edges(data=True)]
fig, ax = plt.subplots(figsize=(10, 10))
pos = nx.spring_layout(G_con_curvatura, seed=42)
edges = nx.draw_networkx_edges(G_con_curvatura, pos, edge_color=curvaturas, edge_cmap=plt.cm.coolwarm, width=2)
nodes = nx.draw_networkx_nodes(G_con_curvatura, pos, node_color='lightgray')
cbar = plt.colorbar(edges, ax=ax)
cbar.set_label('Curvatura de Ricci (calculada con SciPy)')
ax.set_title('Red Cerebral Coloreada por Curvatura', fontsize=16)
plt.show()
```

```python
# === Bloque Final (Versión Corregida): La Simulación del Topoconnectoma Dinámico ===

import os
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import mne
import antropy as ant
from scipy.stats import wasserstein_distance
import imageio

# --- 1. Configuración de la Simulación ---
print("Configurando parámetros de la simulación...")
DURACION_VENTANA = 2.0  # segundos
PASO_VENTANA = 0.5      # segundos
CARPETA_FOTOGRAMAS = 'fotogramas_simulacion'
NOMBRE_VIDEO = 'simulacion_neurotopologica.mp4'
FPS_VIDEO = 5 # Fotogramas por segundo del video final

if not os.path.exists(CARPETA_FOTOGRAMAS):
    os.makedirs(CARPETA_FOTOGRAMAS)

# --- 2. Función de Curvatura ---
def calcular_curvatura(G):
    for u, v in G.edges():
        d_u, d_v = G.degree(u), G.degree(v)
        if d_u == 0 or d_v == 0: continue
        p_u, p_v = np.ones(d_u)/d_u, np.ones(d_v)/d_v
        dist = wasserstein_distance(np.arange(d_u), np.arange(d_v), p_u, p_v)
        G[u][v]['ricciCurvature'] = 1 - dist
    return G

# --- 3. Pre-cálculos ---
G_layout = nx.from_numpy_array(np.abs(np.corrcoef(raw_cleaned.get_data().T, rowvar=False)))
pos = nx.spring_layout(G_layout, seed=42)
sfreq = raw_cleaned.info['sfreq']
n_puntos_ventana = int(DURACION_VENTANA * sfreq)
n_puntos_paso = int(PASO_VENTANA * sfreq)
datos_completos = raw_cleaned.get_data(units='V')
n_fotogramas = int((30 * sfreq - n_puntos_ventana) / n_puntos_paso)
print(f"Se generarán {n_fotogramas} fotogramas. Esto puede tardar varios minutos...")

# --- 4. El Bucle Principal de Generación de Fotogramas ---
lista_fotogramas = []
for i in range(n_fotogramas):
    inicio = i * n_puntos_paso
    fin = inicio + n_puntos_ventana
    datos_ventana = datos_completos[:, inicio:fin]
    
    # a. Calcular Entropía para la ventana
    entropias = [ant.perm_entropy(canal, normalize=True) for canal in datos_ventana]
    
    # b. Calcular Curvatura para la ventana
    matriz_corr_ventana = np.nan_to_num(np.corrcoef(datos_ventana.T, rowvar=False))
    G_ventana = nx.from_numpy_array(np.abs(matriz_corr_ventana))
    G_ventana = calcular_curvatura(G_ventana)
    curvaturas = [d.get('ricciCurvature', 0.0) for u, v, d in G_ventana.edges(data=True)]
    
    # c. Dibujar y guardar el fotograma
    fig, ax = plt.subplots(figsize=(10, 10))
    nodes = nx.draw_networkx_nodes(G_ventana, pos, node_color=entropias, cmap=plt.cm.plasma, node_size=300)
    # LÍNEA CORREGIDA (sin vmin y vmax)
    edges = nx.draw_networkx_edges(G_ventana, pos, edge_color=curvaturas, edge_cmap=plt.cm.coolwarm, width=2)
    
    ax.set_title(f"Topoconnectoma Dinámico | Tiempo: {inicio/sfreq:.1f}s - {fin/sfreq:.1f}s", fontsize=16)
    
    filepath = os.path.join(CARPETA_FOTOGRAMAS, f'frame_{i:04d}.png')
    plt.savefig(filepath)
    plt.close(fig)
    lista_fotogramas.append(filepath)
    
    print(f"Fotograma {i+1}/{n_fotogramas} generado...")

print("\n--- 5. Ensamblando la Película ---")
with imageio.get_writer(NOMBRE_VIDEO, fps=FPS_VIDEO) as writer:
    for filename in lista_fotogramas:
        image = imageio.imread(filename)
        writer.append_data(image)

print(f"\n¡SIMULACIÓN COMPLETADA! El video se ha guardado como '{NOMBRE_VIDEO}' en su carpeta.")
```
