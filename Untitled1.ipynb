```python
# === Bloque 1 (Versión 5 - Simplificada): ===

import mne

print("--- Iniciando el Laboratorio (Modo Simple) ---")

# --- Cargamos los datos ---
filename = 'chb01_01.edf'
raw = mne.io.read_raw_edf(filename, preload=True, verbose=False)

# --- Filtramos los datos ---
raw_filtered = raw.copy()
raw_filtered.filter(l_freq=1.0, h_freq=40.0, verbose=False)

print("\n¡Éxito! Datos cargados y filtrados.")
print("La variable 'raw_filtered' está lista.")
```

```python
# === Bloque 2 (Versión 2 - Robusta): Análisis ICA ===

from mne.preprocessing import ICA

print("Iniciando análisis ICA...")

# Preparamos y entrenamos el algoritmo
ica = ICA(n_components=15, max_iter='auto', random_state=97)
ica.fit(raw_filtered)

# Visualizamos las propiedades de los componentes encontrados
# Esta función es más completa y robusta
print("\nMostrando las propiedades de los componentes de ICA...")
ica.plot_properties(raw_filtered)
```

```python
# === Bloque 3: Visualización de Fuentes de ICA (Método a prueba de errores) ===

print("Mostrando la actividad de las fuentes de ICA a lo largo del tiempo...")
ica.plot_sources(raw_filtered)
```

```python
# === Bloque 4: Exclusión de Artefactos y Limpieza Final ===

# Le decimos al objeto 'ica' el NÚMERO del componente que queremos excluir.
# ICA000 es el componente 0, ICA001 es el 1, y así sucesivamente.
# Vamos a excluir el 0 (parpadeos) y el 1 (posible movimiento ocular).
components_a_excluir = [0, 1]
ica.exclude = components_a_excluir
print(f"Componentes marcados para exclusión: {ica.exclude}")

# Ahora, aplicamos la limpieza.
# Esto crea un nuevo objeto de datos con el ruido de esos componentes ya eliminado.
# 'raw_filtered' es el original, 'raw_cleaned' será el limpio.
raw_cleaned = ica.apply(raw_filtered.copy()) # Aplicamos sobre una copia

print("\nLimpieza completada. Ahora vamos a comparar el ANTES y el DESPUÉS.")

# --- Verificación Visual ---
# Graficamos los datos ANTES de la limpieza para recordar cómo se veían
print("Mostrando datos ANTES de la limpieza...")
raw_filtered.plot(scalings='auto', title='Datos ANTES de la limpieza ICA')

# Graficamos los datos DESPUÉS de la limpieza
print("Mostrando datos DESPUÉS de la limpieza...")
raw_cleaned.plot(scalings='auto', title='Datos DESPUÉS de la limpieza ICA')
```
